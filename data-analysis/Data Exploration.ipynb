Import java run data

```python
runs = ['default']
from glob import glob
import os

csv_format_string = "wealth-distrib-%s-seed-%s.csv"
data_file = os.path.join("..", "java-data", "commit-35cc68e9", csv_format_string)


#print(data_directory)

df = pd.DataFrame()
for i in range(0,20):
    filename = data_file % ('default', i)
    df_new = pd.read_csv(filename)
    df[f"gini-{i}"] = df_new['gini']
    
df
```

```python
plt.plot(df, alpha=0.2, color='red')
plt.xlabel('Tick')
plt.ylabel('Gini index')
```

```python
plt.plot(df, alpha=0.1, color='red')
plt.xlabel('Tick')
plt.ylabel('Gini index')
plt.plot(df_netlogo_1, alpha=0.1, color='blue')
```

```python
df_mean = df.transpose().describe().transpose()
plt.fill_between(df.index, df_mean['mean'] - df_mean['std'], df_mean['mean'] + df_mean['std'], color='blue', alpha=0.4)
plt.plot(df_mean['mean'], color='blue')
plt.xlabel('Tick')
plt.ylabel('Avg Gini index')

#plt.plot(df_mean['mean'] + df_mean['std'], color='blue')
#plt.plot(df_mean['mean'] - df_mean['std'], color='blue')
```

```python
def plot_mean_envelope(df, colour, alpha=0.2):
    df_mean = df.transpose().describe().transpose()
    plt.fill_between(df.index, df_mean['mean'] - df_mean['std'], df_mean['mean'] + df_mean['std'], color=colour, alpha=alpha)
    plt.plot(df_mean['mean'], color=colour)
```

Let's plot the mean value +/- 1 standard deviation (computed for each tick) for both NetLogo and Java:

```python
plot_mean_envelope(df_netlogo_1, 'red')
plot_mean_envelope(df, 'blue')
plt.xlabel('Tick')
plt.ylabel('Avg Gini index')
plt.legend(['NetLogo', 'Java'])
```

We can see that the mean falls within 1 stdev at all times
