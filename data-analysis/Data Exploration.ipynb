Import CSV from netlogo

Get summary stats

```python
df_netlogo_1.describe()
```

Plot of Gini vs time for 20 runs - default values, Netlogo

Lets get summary stats aggregated across runs

```python
run_stats_nl_1 = df_netlogo_1.describe()
summary_stats_nl_1 = run_stats_nl_1.transpose().describe()[["mean", "min", "max"]]
summary_stats_nl_1
```

```python
summary_stats_nl_1.plot(kind='box', ylabel='Gini index')
```
